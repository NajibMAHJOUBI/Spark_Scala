```python
from __future__ import division
import random
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
number_cluster = 4
number_points = 100
xmin, ymin = -10.0, -10.0
xmax, ymax = 10.0, 10.0
radius = 1.0
dx, dy = (xmax-xmin) / number_cluster, (ymax - ymin) / number_cluster
```

```python
cluster_centers = [(xmin + ix * dx, ymin + iy * dy) for ix in range(number_cluster) for iy in range(number_cluster)]
cluster_centers
```

```python
mean = [0, 0]
cov = [[0.5, 0], [0, 0.5]]
```

```python
data = []
for params in cluster_centers:
    x0, y0 = np.random.multivariate_normal(mean, cov, number_points).T
    data.extend(zip(map(lambda x: x + params[0], list(x0)), map(lambda x: x + params[1], list(y0))))
```

```python
data
```

```python
xcoor, ycoor = [], []
for tp in data:
    xcoor.append(tp[0])
    ycoor.append(tp[1])
```

```python
plt.plot(xcoor, ycoor, 'o')
```
